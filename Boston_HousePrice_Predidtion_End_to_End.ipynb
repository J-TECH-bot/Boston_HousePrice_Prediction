```python
#Independent and dependent feature

X= df.drop('Price', axis=1)
y=df['Price']
```

```python
y
```

```python
# Train test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardize the dataset (Because each column having different units and caluculate that different units is difficult or complex so thats why we use use standarddization or normalization for easy calculations)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)#(Not used .fit so taht model not now much about test data set)
```

```python
# We use gradient desent and for that we need come to  gloable minima and to come to the gloable minima we need all indipandent feature in the same scale so that conversion occur easily
X_train
```

```python
#  Model training
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Print  the coefficients and intercepts
print(regressor.coef_)
print(regressor.intercept_)
```

```python
# on which parameters the model has been trained also use for creating model copy with same parameters
regressor.get_params()
```

```python
# prediction with test data
reg_pred = regressor.predict(X_test)
```

```python
reg_pred
```

```python
# Sccator plot for prediction
plt.scatter(y_test, reg_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.show()
```

```python
# Residuals is error between prediction and actual value
residuals = y_test - reg_pred
residuals
```

```python
# plot residuals
# frm that plot we can clearly see that there is normaldistribution with some outliers
sns.displot(residuals, kind='kde')
```

```python
# scatter plot with respect to predication and residuals
# uniform distribution
plt.scatter(reg_pred, residuals)
plt.xlabel('Predicted Price')
plt.ylabel('Residuals')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score

print(mean_squared_error(y_test, reg_pred))#Calculates the average of the squared differences between predicted and actual values.
print(mean_absolute_error(y_test, reg_pred))#Calculates the average of the absolute differences between predicted and actual values.
print(r2_score(y_test, reg_pred))#Represents the proportion of variance in the target variable explained by the model.Ranges from 0 to 1
print(np.sqrt(mean_squared_error(y_test, reg_pred)))#Calculates the square root of the mean of the squared differences between predicted and actual values.`
```
